```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn
from sklearn import datasets
from mpl_toolkits.mplot3d import Axes3D
```

```python
iris = datasets.load_iris()
X = iris.data[:, [0,1]]  # we only take the first two features.
Y = iris.target
uniq=np.unique(Y)
m=3
color=['ro','go','bo']
for i in range(3):
    index=Y==uniq[i]
    plt.plot(X[index,0],X[index,1],color[i])
    
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')   
plt.show()
```

```python
X1=iris.data
Xmean=np.mean(X1,axis=0)
Xnew=X1-Xmean

cov=np.dot(Xnew.T,Xnew)
[value,vector]=np.linalg.eig(cov)

plt.plot(np.arange(0,4),value)
plt.title('eigen value')
plt.ylabel('eigen value')
plt.xlabel('list of number')
plt.show()
```

```python
new_vector=vector[:,[0,1]]
data_new=np.dot(Xnew,new_vector)
print vector
print new_vector 
#print Xnew
```

```python
Y = iris.target
uniq=np.unique(Y)
m=3
color=['ro','go','bo']
for i in range(3):
    index=Y==uniq[i]
    plt.plot(data_new[index,0],data_new[index,1],color[i])
plt.title('PCA of iris dataset')
plt.xlabel('PCA1')
plt.ylabel('PCA2')   
plt.show()
```
